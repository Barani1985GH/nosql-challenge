### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
query = {'BusinessName':'Penang Flavours'}
result = establishments.find_one(query)
result
degree_search = 0.01
long = round(result['geocode']['longitude'],2)
lat = round(result['geocode']['latitude'],2)

# Set longitude search range
search_longitude_1 = long + degree_search
search_longitude_2 = long - degree_search

# Set latitude search range
search_latitude_1 = round((lat + degree_search),2)
search_latitude_2 = round((lat - degree_search),2)


query = {'$and':[
            {
                '$and':[
                    {"geocode.longitude":{'$lt':search_longitude_1}},
                    {"geocode.longitude":{'$gt':search_longitude_2}}
                ]
            },
            {
                '$and':[
                    {"geocode.latitude":{'$lt':search_latitude_1}},
                    {"geocode.latitude":{'$gt':search_latitude_2}}
                ]
            },
            {
                "RatingValue":{'$eq':5}
            }

 ]}

sort = [('scores.Hygiene',1)] 
limit= 5
results = list(establishments.find(query).sort(sort).limit(limit))

# Print the results
for row in results:
    print(row)

# Long, 0.08
# Lat : 51.49
# long range from 0.07  &  0.09
# lat range from 51.48  &  51.5
```

```python
# Convert result to Pandas DataFrame
df = pd.json_normalize(results)
df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match':{'scores.Hygiene':{'$eq':0}}}
# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': {"Local Authority Name": "$LocalAuthorityName"},
                          'count':{'$sum':1}
                          }
                }
# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': {'count': -1 }}
# # Print the number of documents in the result
pipeline = [match_query,group_query,sort_values]
results = list(establishments.aggregate(pipeline))
# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
localAuth_df = pd.json_normalize(results)

# Display the number of rows in the DataFrame
print(len(localAuth_df.index))

# Rename and Reorder columns
localAuth_df.rename(columns={'_id.Local Authority Name':'Local Authority Name'},inplace=True)
localAuth_df = localAuth_df[['Local Authority Name','count']]

# Display the first 10 rows of the DataFrame
localAuth_df.head(10)
```
